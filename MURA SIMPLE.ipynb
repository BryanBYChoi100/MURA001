```python

def modify_csv(path) :
    df = pd.read_csv(path, header=None)
    df = df[df[0].str.contains("XR_ELBOW")].reset_index(drop=True)
    df = df[0].str.replace(pat='MURA-v1.1', repl='MURA_DATA', regex=False)
    df.to_csv(path, index=False, header=False)

print("1")
path = '/content/drive/My Drive/대학원/과제/MURA_DATA/valid_image_paths.csv'
modify_csv(path)
print("2")
path = '/content/drive/My Drive/대학원/과제/MURA_DATA/valid_labeled_studies.csv'
modify_csv(path)
print("3")
path = '/content/drive/My Drive/대학원/과제/MURA_DATA/train_labeled_studies.csv'
modify_csv(path)
print("4")
path = '/content/drive/My Drive/대학원/과제/MURA_DATA/train_image_paths.csv'
modify_csv(path)
```
